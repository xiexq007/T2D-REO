```r
library(dplyr)
library(ggplot2)
library(patchwork)
library(reshape2)
```

```r
raw_data_tpm = read.csv("raw_tpm.csv")
data_long = melt(raw_data_tpm)
ggplot(data_long, aes(x = value)) +
  geom_density(fill = "lightblue", color = "black") +
  facet_wrap(~ variable, ncol = 4, nrow = 3, scales = "free") +
  labs(x = "Gene Expression Level", y = "Density")
```

```r
pseu_tpm = read.csv("pseu_tpm.csv")
data_long = melt(pseu_tpm)
ggplot(data_long, aes(x = value)) +
  geom_density(fill = "#ffe8a3", color = "black") +
  facet_wrap(~ variable, ncol = 4, nrow = 3, scales = "free") +
  labs(x = "Gene Expression Level", y = "Density")
```
